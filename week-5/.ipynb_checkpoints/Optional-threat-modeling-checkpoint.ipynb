## MVC Architecture Diagram

The base architecture for our threat model.

```python
from diagrams import Diagram, Cluster
from diagrams.onprem.client import Users, Client
from diagrams.onprem.network import Internet, Haproxy
from diagrams.programming.framework import Flask
from diagrams.onprem.database import PostgreSQL
from diagrams.aws.network import ELB
from IPython.display import Image

with Diagram("Flask Web Platform with MVC Architecture", show=False, outformat="png", filename="flask_mvc_arch"):
    user = Users("User")  # End-user
    browser = Client("Browser")  # View layer
    internet = Internet("Internet")  # Network
    lb = ELB("Load Balancer")  # Traffic distribution

    with Cluster("Web Platform (MVC)"):
        view = browser
        nginx = Haproxy("Nginx (Web Server)")  # View serving
        controller = Flask("Flask App (Controller)")  # Logic
        model = PostgreSQL("PostgreSQL (Model)")  # Data
        view >> nginx >> controller >> model

    user >> view >> internet >> lb >> nginx

Image(filename="flask_mvc_arch.png")
```

## Threat Model for Flask MVC Platform

Using STRIDE methodology for threat identification.

### Sequence Diagram

Shows user interaction flow.

```python
with Diagram("Sequence Diagram: User Interaction with MVC Platform", show=False, outformat="png", filename="mvc_sequence"):
    user = Users("User")
    browser = Client("Browser")
    lb = ELB("Load Balancer")
    nginx = Haproxy("Nginx (Web Server)")
    flask = Flask("Flask App (Controller)")
    db = PostgreSQL("PostgreSQL (Model)")

    user >> browser >> lb >> nginx >> flask >> db
    db >> flask >> nginx >> lb >> browser >> user

Image(filename="mvc_sequence.png")
```

### Data Flow Diagram (DFD)

Illustrates data movement and trust boundaries.

```python
with Diagram("Data Flow Diagram: Flask MVC Platform", show=False, outformat="png", filename="mvc_dfd"):
    user = Users("User")
    internet = Internet("Internet")

    with Cluster("External Boundary"):
        browser = Client("Browser")

    with Cluster("Web Platform Trust Boundary"):
        lb = ELB("Load Balancer")
        nginx = Haproxy("Nginx (Web Server)")
        flask = Flask("Flask App (Controller)")
        db = PostgreSQL("PostgreSQL (Model)")

        browser >> internet >> lb >> nginx >> flask >> db
        db >> flask >> nginx >> lb >> internet >> browser

Image(filename="mvc_dfd.png")
```
