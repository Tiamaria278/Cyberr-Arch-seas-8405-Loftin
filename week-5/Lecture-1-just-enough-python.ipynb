# Section 4: Introduction to `diagrams`

`diagrams` creates architecture visuals.

## Simple Diagram
User to server:

````python
from graphviz import Digraph

dot = Digraph()
dot.node('A', 'Web Server')
dot.node('B', 'App Server')
dot.node('C', 'Database')
dot.edges(['AB', 'BC'])

dot  # just return the object to render in Jupyter
````

````python
from graphviz import Digraph
from IPython.display import Image
dot = Digraph("Web_Service", format="png")
dot.attr(label="Web Service Architecture", fontsize="16", labelloc="t")
dot.node("User", "User (Browser)")
dot.node("DNS", "DNS Resolver")
dot.node("LB", "Load Balancer")
dot.node("WS1", "Web Server 1 (Nginx)")
dot.node("WS2", "Web Server 2 (Apache)")
dot.node("Static", "Static Content")
dot.node("App", "Application Server")
dot.edge("User", "DNS")
dot.edge("DNS", "LB")
dot.edge("LB", "WS1")
dot.edge("LB", "WS2")
dot.edge("WS1", "Static")
dot.edge("WS2", "App")
filename = "arch/Web_Service_diagram"
dot.render(filename, format="png", cleanup=False)
Image(filename=f"{filename}.png")
````

````python
from diagrams import Diagram
from diagrams.onprem.client import User
from diagrams.onprem.compute import Server
from IPython.display import Image

with Diagram("Client-Server", show=False, outformat="png", filename="simple_diagram"):
    user = User("User")
    server = Server("Server")
    user >> server

Image(filename="simple_diagram.png")
````

- `with Diagram()`: Defines the diagram.
- `>>`: Connects components.

# Section 5: Introduction to `tabulate`

`tabulate` formats data as tables.

## Simple Table
Cybersecurity tools:

````python
from tabulate import tabulate

data = [
    ["Firewall", "Blocks unauthorized access"],
    ["VPN", "Secures remote connections"],
    ["IDS", "Detects intrusions"]
]

headers = ["Tool", "Purpose"]

print(tabulate(data, headers=headers, tablefmt="fancy_grid"))
````

# Section 7: Advanced Python Concepts

Let's explore `if-else` and advanced `for` loops.

## If-Else Statements
Make decisions:

````python
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C")
````

Use cases: Check security conditions (e.g., "Is the port open?").

## Advanced For Loops
Process data with conditions:

````python
ports = [22, 80, 443, 8080]
for port in ports:
    if port == 22:
        print(f"Port {port}: SSH - Secure!")
    elif port == 80:
        print(f"Port {port}: HTTP - Warning: Unencrypted!")
    else:
        print(f"Port {port}: Other service.")
````

Combine with lists or dictionaries for complex tasks.

## Example: Security Check
Check open ports:

````python
open_ports = [80, 443]
for port in open_ports:
    status = "Secure" if port == 443 else "Unsecure"
    print(f"Port {port} is {status}.")
````
